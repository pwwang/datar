```python
# https://dplyr.tidyverse.org/reference/setops.html

from plyrda.data import mtcars
from plyrda.all import *
```

```python
first = mtcars >> slice(f[:20])
second = mtcars >> slice(f[9:33])

intersect(first, second) # or first >> intersect(second)
```

```python
union(first, second)
```

```python
setdiff(first, second)
```

```python
setdiff(second, first)
```

```python
union_all(first, second)
```

```python
setequal(mtcars, mtcars >> slice(f[::-1]))
```
