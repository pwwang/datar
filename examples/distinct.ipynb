```python
# https://dplyr.tidyverse.org/reference/distinct.html
from random import choices
from plyrda.data import starwars
from plyrda.all import *
```

```python
df = tibble(
  x=choices(range(10), k=100),
  y=choices(range(10), k=100)
)
nrow(df)
```

```python
nrow(distinct(df))
```

```python
df >> distinct(f.x, f.y) >> nrow()
```

```python
df >> distinct(f.x)
```

```python
df >> distinct(f.y)
```

```python
df >> distinct(f.x, _keep_all=True)
```

```python
df >> distinct(f.y, _keep_all=True)
```

```python
df >> distinct(diff=abs(f.x-f.y))
```

```python
starwars >> distinct(across(contains("color")))
```

```python
df = tibble(
  g=[1, 1, 2, 2],
  x=[1, 1, 2, 1]
) >> group_by(f.g)

df >> distinct(f.x) >> ungroup()
```
